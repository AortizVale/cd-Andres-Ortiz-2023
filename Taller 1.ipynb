```python
import pandas as pd
import numpy as np
```

```python
#Insertar datos
x = np.array([75, 35, 105, 50, 85, 72, 120, 90])
y = np.array([200, 120, 210, 155, 300, 240, 245, 250])
```

```python
#se reescribe la ecuación como y = Ap, donde A = [[x 1]] y p = [[m], [c]].
A = np.vstack([x, np.ones(len(x))]).T
A
```

```python
#Se busca la pendiente y el intercepto
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
m, c
```

```python
#Mostrar la ecuación
print("y = "+str(m)+"x + "+str(c))
```

```python
#Mostrar gráfica
import matplotlib.pyplot as plt
_ = plt.plot(x, y, 'o', label='Datos', markersize=10)
_ = plt.plot(x, m*x + c, 'r', label='Regresión lineal')
plt.xlabel("Área")
plt.ylabel("Precio")
_ = plt.legend()
plt.show()
```
